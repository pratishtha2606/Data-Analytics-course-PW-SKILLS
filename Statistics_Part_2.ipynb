# **Statistics Part 2** **Theory**

1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine whether a certain assumption (hypothesis) about the population is likely to be true.

We begin by forming two hypotheses:

Null hypothesis (H‚ÇÄ): Assumes no effect or no difference.

Alternative hypothesis (H‚ÇÅ or Ha): Assumes an effect or difference exists.

Based on sample data, we use statistical tests to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative.




2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

The null hypothesis (H‚ÇÄ) is a statement of no effect, no difference, or status quo. It is the hypothesis we test against.

The alternative hypothesis (H‚ÇÅ or Ha) is a statement that contradicts the null hypothesis and represents what we aim to prove.

Difference:

H‚ÇÄ is assumed true until proven otherwise.

H‚ÇÅ is accepted only if there is sufficient evidence against H‚ÇÄ.

Example:

H‚ÇÄ: The average battery life is 10 hours.

H‚ÇÅ: The average battery life is not 10 hours.

3. What is the significance level in hypothesis testing, and why is it important?

The significance level (denoted as Œ±) is the probability of rejecting the null hypothesis when it is actually true. It defines the threshold for how extreme the test results must be to reject H‚ÇÄ.

Common values: 0.05 (5%), 0.01 (1%)

It is important because it controls the Type I error rate‚Äîthe risk of making a false positive conclusion.

Example:
If Œ± = 0.05, we are willing to accept a 5% chance of incorrectly rejecting H‚ÇÄ.



4. What does a P-value represent in hypothesis testing?

The P-value is the probability of obtaining a test statistic equal to or more extreme than the observed value, assuming that the null hypothesis is true.

It tells us how likely the observed data is under the assumption that H‚ÇÄ is correct.

A smaller P-value indicates stronger evidence against H‚ÇÄ.

5. How do you interpret the P-value in hypothesis testing?

If P-value ‚â§ Œ± (significance level) ‚Üí Reject the null hypothesis.

If P-value > Œ± ‚Üí Fail to reject the null hypothesis.

Example:

If the P-value is 0.03 and Œ± = 0.05, we reject H‚ÇÄ because the result is statistically significant.

If P-value is 0.08 and Œ± = 0.05, we do not reject H‚ÇÄ.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Type 1 Error (Œ±): Rejecting the null hypothesis when it is actually true.

Example: Saying a drug works when it actually doesn‚Äôt.

Type 2 Error (Œ≤): Failing to reject the null hypothesis when it is actually false.

Example: Saying a drug doesn't work when it actually does.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

One-tailed test: Tests for the possibility of the effect in only one direction.

H‚ÇÅ: Œº > 50 or H‚ÇÅ: Œº < 50

Two-tailed test: Tests for the possibility of the effect in both directions.

H‚ÇÅ: Œº ‚â† 50

Key difference:

One-tailed is more powerful if direction is known.

Two-tailed is safer when we‚Äôre unsure of direction.



8. What is the Z-test, and when is it used in hypothesis testing?

A Z-test is used when:

The sample size is large (n > 30)

Population standard deviation is known

Data follows a normal distribution

It is commonly used to compare:

Sample mean vs population mean

Proportions

Example: Testing if the average height of students in a college differs from the national average.

You're testing means from a small sample

19. What is the F-distribution, and when is it used in hypothesis testing?

20. What is an ANOVA test, and what are its assumptions?

ANOVA (Analysis of Variance) compares means of three or more groups to see if at least one group mean is different.

Hypotheses:

H‚ÇÄ: All group means are equal

H‚ÇÅ: At least one group mean is different

Assumptions:

Independence of observations

Normality of data within groups

Equal variances across groups (homogeneity)

21. What are the different types of ANOVA tests?

One-way ANOVA: One independent variable with 3+ levels/groups

Two-way ANOVA: Two independent variables, can also test interaction effect

Repeated Measures ANOVA: Same subjects tested multiple times (within-subject design)

MANOVA: Multivariate ANOVA for multiple dependent variables

Use in hypothesis testing:

It helps in model evaluation, especially in regression and ANOVA.

 2. Generate a discrete uniform distribution and plot PMF


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a discrete uniform distribution (1 to 6)
data = np.random.randint(1, 7, size=1000)

# Calculate the unique values and their frequencies
values, counts = np.unique(data, return_counts=True)

# Calculate the Probability Mass Function (PMF)
pmf = counts / counts.sum()

# Plot the PMF
plt.stem(values, pmf)  # Removed use_line_collection
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.ylim(0, max(pmf) + 0.05)
plt.grid(True)
plt.show()
```

Q9-Write a Python function to calculate Z-scores from a dataset and plot them

```python
def plot_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=30, edgecolor='black')
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.title("Histogram of Z-scores")
    plt.grid(True)
    plt.show()

# Example usage
sample_data = np.random.normal(loc=100, scale=15, size=1000)
plot_z_scores(sample_data)
```

Q10- Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
def simulate_clt(sample_size=30, num_samples=1000):
    sample_means = []

    for _ in range(num_samples):
        # Use exponential distribution to show CLT (non-normal)
        sample = np.random.exponential(scale=2.0, size=sample_size)
        sample_means.append(np.mean(sample))

    # Plot histogram of sample means
    plt.hist(sample_means, bins=30, edgecolor='black')
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.title(f"Central Limit Theorem (Sample Size = {sample_size})")
    plt.grid(True)
    plt.show()

# Run the simulation
simulate_clt()
```

Q1. Perform a Z-test comparing sample mean to population mean and interpret results

```python
from scipy import stats
import numpy as np

# Sample data
sample = np.random.normal(loc=52, scale=10, size=50)
sample_mean = np.mean(sample)
pop_mean = 50
pop_std = 10
n = len(sample)

# Z-test formula
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-value:", z)
print("P-value:", p_value)
```

Q4. Two-tailed Z-test with decision region visualization

```python
import matplotlib.pyplot as plt

z_score = z
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.axvline(z_critical, color='red', linestyle='--', label='Critical Region')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.axvline(z_score, color='blue', label='Z Score')
plt.title("Two-Tailed Z-Test Decision Region")
plt.legend()
plt.grid(True)
plt.show()
```

Q5. Function to calculate & visualize Type 1 and Type 2 errors

```python
def visualize_errors(mu0, mu1, sigma, n, alpha=0.05):
    z_critical = stats.norm.ppf(1 - alpha)
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    plt.plot(x, stats.norm.pdf(x, mu0, sigma/np.sqrt(n)), label='H0 Distribution')
    plt.plot(x, stats.norm.pdf(x, mu1, sigma/np.sqrt(n)), label='H1 Distribution')

    cutoff = mu0 + z_critical * (sigma / np.sqrt(n))
    plt.axvline(cutoff, color='red', linestyle='--', label='Critical Value')

    plt.fill_between(x, 0, stats.norm.pdf(x, mu0, sigma/np.sqrt(n)), where=(x >= cutoff), alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, stats.norm.pdf(x, mu1, sigma/np.sqrt(n)), where=(x < cutoff), alpha=0.3, label='Type II Error')

    plt.legend()
    plt.title("Type I and Type II Error Visualization")
    plt.grid(True)
    plt.show()

visualize_errors(mu0=100, mu1=105, sigma=15, n=30)
```

 Q9. Function to calculate confidence interval for sample mean

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return interval

data = np.random.normal(50, 10, 40)
print("Confidence Interval:", confidence_interval(data))
```

 Q10. Calculate margin of error for confidence level

```python
data = np.random.normal(70, 12, 100)
confidence = 0.95

sem = stats.sem(data)
margin_error = stats.t.ppf((1 + confidence) / 2, len(data)-1) * sem

print("Margin of Error:", margin_error)
```

 Q11. Implement Bayesian inference using Bayes‚Äô Theorem

```python
# P(A|B) = [P(B|A) * P(A)] / P(B)

def bayes_theorem(p_b_given_a, p_a, p_b):
    return (p_b_given_a * p_a) / p_b

# Example: medical test
p_disease = 0.01
p_positive_given_disease = 0.99
p_positive = 0.05

p_disease_given_positive = bayes_theorem(p_positive_given_disease, p_disease, p_positive)
print("P(Disease | Positive Test):", p_disease_given_positive)
```
