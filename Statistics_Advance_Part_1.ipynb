13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:

Importance:

Allows using normal distribution for inferential statistics, even when population is not normal.

Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Generate random integers from 1 to 6 (like a fair dice)
data = randint.rvs(1, 7, size=1000)

# Compute PMF
values, counts = np.unique(data, return_counts=True)
probabilities = counts / counts.sum()

# Plot PMF
plt.bar(values, probabilities, color='skyblue', edgecolor='black')
plt.title("PMF of Discrete Uniform Distribution (Dice Roll)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
from scipy.stats import bernoulli

# Parameters
p = 0.6
x = [0, 1]

# Calculate PDF
pdf = bernoulli.pmf(x, p)

# Print PDF values
for xi, pi in zip(x, pdf):
    print(f"P(X = {xi}) = {pi}")
```

Q19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample from population
sample = np.random.normal(loc=102, scale=15, size=40)
sample_mean = np.mean(sample)

# Known population mean and std
pop_mean = 100
pop_std = 15
n = len(sample)

# Z-statistic
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")
```

Q20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

# Sample data
data = np.random.normal(loc=100, scale=12, size=40)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval
z = norm.ppf(0.975)
margin_error = z * (std / np.sqrt(n))

ci_lower = mean - margin_error
ci_upper = mean + margin_error

print(f"Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q21- Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
from scipy.stats import norm

# Step 1: Generate normal data
data = np.random.normal(loc=70, scale=10, size=50)  # mean=70, std=10, n=50

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Step 3: Calculate 95% confidence interval
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)
margin_error = z_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

# Generate normal dataset
data = np.random.normal(loc=70, scale=10, size=100)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Output
print(f"Range = {data_range:.2f}")
print(f"Interquartile Range (IQR) = {iqr:.2f}")
```

Q27. Implement Z-score normalization on a dataset and visualize its transformation.
